```python
def diagnosis_proportion(data):
    return data["diagnosis"].value_counts() / len(data)

train_set, test_set = train_test_split(cancer, test_size=0.2, random_state=42)

compare_props = pd.DataFrame({
    "Overall %": diagnosis_proportion(cancer),
    "Stratified %": diagnosis_proportion(strat_test_set),
    "Random %": diagnosis_proportion(test_set),
}).sort_index()
compare_props.index.name = "Diagnosis Category"
compare_props["Strat. Error %"] = (compare_props["Stratified %"] /
                                   compare_props["Overall %"] - 1)
compare_props["Rand. Error %"] = (compare_props["Random %"] /
                                  compare_props["Overall %"] - 1)
(compare_props * 100).round(2)
```

```python
X_train = train_set.drop("diagnosis", axis = 1)
y_train = train_set["diagnosis"].copy()
```

```python
y_train_b = (y_train == 0)
y_train_b.count()
```

```python
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier()
sgd_clf.fit(X_train,y_train_b)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
standard_scaler = StandardScaler()
X_train_scaled = standard_scaler.fit_transform(X_train.astype('float64'))
cross_val_score(sgd_clf,X_train_scaled,y_train_b,cv = 3, scoring='accuracy')
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import ConfusionMatrixDisplay
y_train_pred = cross_val_predict(sgd_clf,X_train_scaled,y_train_b,cv=3)
ConfusionMatrixDisplay.from_predictions(y_train_b,y_train_pred)
plt.show()
```

```python

ConfusionMatrixDisplay.from_predictions(y_train_b,y_train_pred,normalize="true", values_format=".0%")
plt.show()
```

```python
from sklearn.model_selection import cross_val_predict
y_scores = cross_val_predict(sgd_clf, X_train,y_train_b,cv=3,method='decision_function')
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_train_b,y_scores)
```

```python
from sklearn.metrics import precision_recall_curve
precisions, recalls, thresholds = precision_recall_curve(y_train_b, y_scores)
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import roc_auc_score
models = {
    'sgd_clf':SGDClassifier(random_state = 42),
    'svc':SVC(random_state = 42),
    'log_reg':LogisticRegression(max_iter = 10000),
    'rand_forest':RandomForestClassifier(random_state = 42)
}
for model,clf in models.items():
  print(f"Training {model}...")
  standard_scaler = StandardScaler()
  X_train_scaled = standard_scaler.fit_transform(X_train.astype('float64'))
  clf.fit(X_train_scaled,y_train)
  cross_val = cross_val_score(clf,X_train_scaled,y_train,cv = 3, scoring='accuracy')
  y_train_pred = cross_val_predict(clf,X_train_scaled,y_train,cv=3)
  ConfusionMatrixDisplay.from_predictions(y_train,y_train_pred,normalize="true", values_format=".0%")
  plt.show()
  if(model == 'rand_forest'):
    y_scores = cross_val_predict(clf, X_train,y_train,cv=3)
  else:
    y_scores = cross_val_predict(clf, X_train,y_train,cv=3,method='decision_function')
  roc = roc_auc_score(y_train,y_scores)
  print(roc)
```
