```python
from sklearn.model_selection import train_test_split
strat_train_set, strat_test_set = train_test_split(cancer, test_size = 0.2, stratify=cancer["diagnosis"], random_state = 42)
```

```python
strat_test_set["diagnosis"].value_counts() / len(strat_test_set)
```

```python
def diagnosis_proportion(data):
    return data["diagnosis"].value_counts() / len(data)

train_set, test_set = train_test_split(cancer, test_size=0.2, random_state=42)

compare_props = pd.DataFrame({
    "Overall %": diagnosis_proportion(cancer),
    "Stratified %": diagnosis_proportion(strat_test_set),
    "Random %": diagnosis_proportion(test_set),
}).sort_index()
compare_props.index.name = "Diagnosis Category"
compare_props["Strat. Error %"] = (compare_props["Stratified %"] /
                                   compare_props["Overall %"] - 1)
compare_props["Rand. Error %"] = (compare_props["Random %"] /
                                  compare_props["Overall %"] - 1)
(compare_props * 100).round(2)
```
