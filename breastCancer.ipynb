```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
num_pipeline = make_pipeline(SimpleImputer(strategy = 'median'), StandardScaler())
```

```python
# Define the preprocessing pipeline
preprocess_pipeline = ColumnTransformer([
    ('num', num_pipeline, cancer_num.columns)
])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import ConfusionMatrixDisplay, roc_auc_score
import matplotlib.pyplot as plt
import pandas as pd


# Define your models
models = {
    'sgd_clf': SGDClassifier(random_state=42),
    'svc': SVC(random_state=42),
    'log_reg': LogisticRegression(max_iter=1000, random_state=42),
    'rand_forest': RandomForestClassifier(random_state=42)
}

# Iterate over each model, create a pipeline, and evaluate
for model_name, model in models.items():
    print(f"Training {model_name}...")

    # Create a pipeline with the preprocessing and the model
    full_pipeline = Pipeline([
        ('preprocessing', preprocess_pipeline),
        ('classifier', model)
    ])

    # Perform cross-validation
    cross_val = cross_val_score(full_pipeline, train_set, cancer_labels, cv=3, scoring='accuracy')
    print(f"Cross-validation accuracy for {model_name}: {cross_val.mean():.2f}")

    # Generate cross-validated predictions
    y_train_pred = cross_val_predict(full_pipeline, train_set, cancer_labels, cv=3)

    # Display the confusion matrix
    ConfusionMatrixDisplay.from_predictions(cancer_labels, y_train_pred, normalize="true", values_format=".0%")
    plt.title(f'Confusion Matrix for {model_name}')
    plt.show()

    # Generate scores for ROC AUC
    if model_name == 'rand_forest':
        y_scores = cross_val_predict(full_pipeline, train_set, cancer_labels, cv=3, method='predict_proba')[:, 1]
    else:
        y_scores = cross_val_predict(full_pipeline, train_set, cancer_labels, cv=3, method='decision_function')

    # Calculate and print ROC AUC
    roc = roc_auc_score(cancer_labels, y_scores)
    print(f"ROC AUC for {model_name}: {roc:.2f}\n")
```

```python
best_model = Pipeline([
    ('preprocessing', preprocess_pipeline),
    ('classifier', SVC(random_state=42))  # Ensure probability=True for predict_proba method
])
best_model.fit(train_set, cancer_labels)
```

```python
X_test = test_set.drop("diagnosis", axis=1)
y_test = test_set["diagnosis"].copy()
```

```python
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay, classification_report, roc_auc_score

# Predict labels for the test set
y_test_pred = best_model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_test_pred)
print(f"Test Accuracy: {accuracy:.2f}")

# Display the confusion matrix
ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred, normalize='true', values_format='.0%')
plt.title('Confusion Matrix for SVC on Test Set')
plt.show()

# Classification report
report = classification_report(y_test, y_test_pred)
print(report)

# Calculate ROC AUC (if applicable)
y_scores = best_model.decision_function(X_test)  # or use predict_proba if probability=True

roc_auc = roc_auc_score(y_test, y_scores)
print(f"ROC AUC Score: {roc_auc:.2f}")
```
